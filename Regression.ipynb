```python
import numpy as np
```

```python
def multivariateRegression(X, y):
    #goal - find w*
    
    X_trans = np.transpose(X)

    print(np.matmul(matInverse(np.matmul(X_trans, X)), X_trans.dot(y)))
    return np.matmul(matInverse(np.matmul(X_trans, X)), X_trans.dot(y))

def matInverse(mat):

    det = mat[0][0]*mat[1][1] - mat[0][1]*mat[1][0]
    flip = np.array([[mat[1][1], -mat[0][1]], [-mat[1][0], mat[0][0]]])
    return (1/det)*flip

def buildX(funcs, x):
    #create X
    X_arr = []

    for datapoint in x:
        row = [] 
        for func in funcs:
            
            row.append(func(datapoint))
            
        X_arr.append(row)
        
    return np.array(X_arr)

def genNonLinearModel(wStar, funcs):
    model = []
    
    i = 0
    
    D = len(wStar)
    
    while i < D:
        model.append([wStar[i], funcs[i]])
        i += 1
        
    return model


        
    
def buildModel(x, y, funcs=[], linear=False):
    if not linear:
        X = buildX(funcs, x)

        w_star = multivariateRegression(X, y)

        return genNonLinearModel(w_star, funcs)
    else:
        return multivariateRegression(x, y)

        
        
#multivariate regression can also be used to build non-linear models
def applyNonLinearModel(x, model):
    res = []
    
    for datapoint in x:
        row = 0
        for sec in model:
            row += sec[0]*sec[1](datapoint) #w_i * f_i(x)
            
        res.append(row)
        
    return np.array(res).T

def applyModel(x, model):
    res = []
    
    for datapoint in x:
        row = 0
        for sec in model: #apply the weights to each data point as a linear combination
            row += sec*datapoint
            
        res.append(row)
        
    return np.array(res).T
    
#example functions
def constant(x, k=1):
    return k

def linear(x, k=1):
    return k*x

#Example
    
X_test = np.array([[-2,-1,0,1,2],[5,4,3,2,1]]).T
y_test = np.array([-6.2,-2.6,0.5,2.7,5.7]).T

funcs_test = [constant, linear]

X_data = np.array([4,1,2,6,7]).T


#print(applyNonLinearModel(X_data, buildModel(X_test, y_test, funcs_test)))
print(applyModel(X_data, buildModel(X_test, y_test, linear=True)))
    
    
```
